```python
import math 
# Function to calculate factorial using a loop
def factorial(k):
    """
    the function calculate the factorial of a non-negative integer n.
    -Input:
    n-The number for which the factorial is to be calculated.
    -Output:
    The factorial of n.
    Raises:
    ValueError: If n is negative.
    """
    if k < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    result = 1
    for i in range(1, k + 1):
        result *= i   
    return result
#testing the function
k=5
factorial(k)

```

```python
import numpy as np
import scipy 
import math 

# Function to manually calculate Poisson probability

def poisson_probability(lam, k):
    """
    Calculates the probability of observing exactly k events given the average rate of events (lam).

    Input:
    [lam]: Average rate of occurrences. Can't handle negative numbers.
    [k]: Number of occurrences to calculate the probability for. can't handle negative numbers .

    Output:
    Probability of observing exactly k events.

    ERROR:
   error If lambda or k are invalid, if else- just print the value
    """
    # Calculate Poisson probability using the formula
    poisson_probability = (lam**k * math.exp(-lam)) / (factorial(k))
    return poisson_probability

# Test case parameters
lam = -1
k = 3
    
# Check for valid inputs manually
if lam < 0:
    raise ValueError ("Error: Lambda must be non-negative.")
elif k < 0:
    raise ValueError ("Error: k must be a non-negative integer.")

# Calculate the probability
result = poisson_probability (lam, k)
print(f"The probability of observing {k} events for lambda = {lam} is {result:.4f}")
```

## plot poisson distribution 

```python
def plot_poisson_distribution(lam, k_range=20):
    """
    plots the Poisson distribution for a given lambda (lam) over a range of k values.

    input arguments:
    lam: The average rate of occurrences (lambda). Must be >= 0.
    k_range: The range ofs value of k to consider. 

    output:
      Bar chart of the Poisson distribution.
    """
    # Generate k values
    k_range = range(0, k_range + 1)

    # Calculate probabilities for each k value
    probabilities = [poisson_probability(lam, k) for k in k_range]

    # Plotting the distribution
    plt.figure(figsize=(8, 6))
    plt.bar(k_range, probabilities, width=0.5, alpha=0.5, color='green', edgecolor='purple')
    plt.title(f"Poisson Distribution (\u03bb = {lam})", fontsize=16)
    plt.xlabel("Number of Events (k)", fontsize=10)
    plt.ylabel("Probability", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.show()

# distribution as asked in the exercise 
plot_poisson_distribution(lam=4, k_range=15)
```

## Calculating the Cumulative Probability

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def factorial(k):
    """
    Calculates the factorial of a non-negative integer k.

    Input arguments:
    k (int): The number for which the factorial is to be calculated. Must be non-negative.

    Output:
    int: The factorial of k.

    Raises:
    ValueError: If k is negative.
    """
    if k < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    result = 1
    for i in range(1, k + 1):
        result *= i
    return result

def poisson_probability(lam, k):
    """
    Calculates the probability of observing exactly k events in a Poisson distribution.

    Input arguments:
    lam (float): The average rate of occurrences (lambda). Must be non-negative.
    k (int): The number of occurrences to calculate the probability for. Must be non-negative.

    Output:
    float: The probability of observing exactly k events.

    Raises:
    ValueError: If lambda or k is invalid (negative values).
    """
    poisson_probability = (lam**k * math.exp(-lam)) / (factorial(k))
    return poisson_probability

def cumulative_poisson_probability(lam, n):
    """
    Calculates the cumulative probability of observing up to n events in a Poisson distribution.

    Input arguments:
    lam (float): The average rate of occurrences (lambda). Must be non-negative.
    n (int): The maximum number of occurrences to calculate the cumulative probability for. Must be non-negative.

    Output:
    float: The cumulative probability of observing up to n events.

    Raises:
    ValueError: If lambda or n is invalid (negative values).
    """
    if lam < 0:
        raise ValueError("Lambda must be non-negative.")
    if n < 0:
        raise ValueError("n must be a non-negative integer.")

    cumulative_probability = 0
    for k in range(0, n + 1):
        cumulative_probability += poisson_probability(lam, k)
    return cumulative_probability

def plot_cumulative_poisson_distribution(lam, k_range=20):
    """
    Plots the cumulative Poisson distribution for a given lambda (lam) over a range of k values.

    Input arguments:
    lam (float): The average rate of occurrences (lambda). Must be >= 0.
    k_range (int, optional): The maximum value of k to consider.

    Output:
    A bar chart of the cumulative Poisson distribution.
    """
    # Generate k values
    k_values = range(0, k_range + 1)

    # Calculate cumulative probabilities
    cumulative_probabilities = [cumulative_poisson_probability(lam, k) for k in k_values]

    # Plotting the cumulative distribution
    plt.figure(figsize=(8, 6))
    plt.bar(k_values, cumulative_probabilities, color='purple', edgecolor='black', alpha=0.7)
    plt.title(f"Cumulative Poisson Distribution (Œª = {lam})", fontsize=16)
    plt.xlabel("Number of Events (k)", fontsize=12)
    plt.ylabel("Cumulative Probability", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.show()

# Example usage for cumulative distribution plot
plot_cumulative_poisson_distribution(lam=3, k_range=5)
print(f"Cumulative probability of observing up to 5 events for Œª=3: {cumulative_prob_3:.4f}")
```

## spike generator 2.0- introducing the physiological feature of refractory period

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_poi_spikes_with_refractory(r, dt, total_time, refractory_period=0.001, recovery_rate_factor=5, plot=False):
    """
    Generate a Poisson spike train with a refractory period and linear recovery.
    
    Parameters:
    r : Base firing rate (spikes per second).
    dt: Time step (in seconds).
    total_time: Total simulation time (in seconds).
    refractory_period: Refractory period (in seconds). Default is 1 ms (0.001 seconds).
    recovery_rate_factor: Time constant for linear recovery of the firing rate.
    
    
    output:
        - np.ndarray: Binary spike train (0 or 1 for each time bin).
        - np.ndarray: Corresponding time vector.
    """
    # Input validation
    if r < 0:
        raise ValueError("Rate parameter 'r' must be non-negative.")
    if dt <= 0:
        raise ValueError("Time step 'dt' must be greater than zero.")
    if total_time < 0:
        raise ValueError("Total time 'total_time' must be non-negative.")
    if refractory_period < 0:
        raise ValueError("Refractory period must be non-negative.")

    total_size = int(total_time / dt)  # Number of time bins
    spike_train = np.zeros(total_size, dtype=int)  # Initialize spike train
    time_vector = np.linspace(0, total_time, total_size, endpoint=False)  # Time vector
    
    # Initialize recovery state
    next_spike_allowed = 0  # Time of next allowed spike due to refractory period

    for t_idx, t in enumerate(time_vector):
        # Recovery rate: r_recover = r * (t - last_spike) / recovery_rate_factor
        if t < next_spike_allowed:
            recovery_rate = 0  # Refractory period, no spike allowed
        else:
            recovery_rate = r * (t - next_spike_allowed) / recovery_rate_factor
        
        # Generate a spike based on the current recovery rate
        if np.random.rand() < recovery_rate * dt:
            spike_train[t_idx] = 1  # Spike occurs
            next_spike_allowed = t + refractory_period  # Update refractory window

    # Plot the spike train
    if plot:
        plt.figure(figsize=(15, 3))
        plt.step(time_vector, spike_train, where='post', color='olive')
        plt.xlabel('Time (s)')
        plt.ylabel('Spikes (0 or 1)')
        plt.title(f'Poisson Spike Train with Refractory Period ({refractory_period * 1000:.1f} ms)')
        plt.show()

    return spike_train, time_vector


# Parameters
r = 30  # Spikes per second
dt = 0.001  # Time step in seconds
total_time = 90  # Simulation duration in seconds
refractory_period = 0.001  # 1 ms refractory period
recovery_rate_factor = 5  # Linear recovery time factor

# Generate spike train with refractory period
spike_train, time_vector = generate_poi_spikes_with_refractory(
    r, dt, total_time, refractory_period=refractory_period, recovery_rate_factor=recovery_rate_factor, plot=True
)
```

## analyzing the spike train

```python
# Analyze the Spike Train
"""
Analysis:
    -Time Interval Histogram (TIH): Shows ISI distribution, indicating timing variability between spikes.
    -Survivor Function: Probability that ISI exceeds a given value, illustrating spike timing decay.
    -Hazard Function: Instantaneous spiking rate, showing likelihood of spiking as time progresses.
    """

#Time Interval Histogram (TIH)
spike_times = time_vector[spike_train == 1]  # Extract spike times
isi = np.diff(spike_times)  # Inter-spike intervals

# Plot the TIH
plt.figure(figsize=(8, 5))
plt.hist(isi, bins=50, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Inter-Spike Interval (s)')
plt.ylabel('Frequency')
plt.title('Time Interval Histogram (TIH)')
plt.show()

#Survivor Function
sorted_isi = np.sort(isi)
survivor_prob = 1 - np.arange(1, len(sorted_isi) + 1) / len(sorted_isi)

plt.figure(figsize=(8, 5))
plt.step(sorted_isi, survivor_prob, where='post', color='green')
plt.xlabel('Inter-Spike Interval (s)')
plt.ylabel('Survivor Function (P(ISI > t))')
plt.title('Survivor Function')
plt.show()

#Hazard Function
hazard_function = survivor_prob / sorted_isi

plt.figure(figsize=(8, 5))
plt.plot(sorted_isi, hazard_function, color='orange')
plt.xlabel('Inter-Spike Interval (s)')
plt.ylabel('Hazard Function (\u03bb(t))')
plt.title('Hazard Function')
plt.ylim(0, max(hazard_function[~np.isnan(hazard_function)]))  
plt.show()
```

## bursty neuron √Ä LA fast spiking neuron

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_poi_spikes_with_refractory(r, dt, total_time, refractory_period=0.001, recovery_rate_factor=5, burst_prob=0, burst_duration=0.01, plot=False):
    """
    Generate a Poisson spike train with a refractory period, linear recovery, and optional bursty behavior.

    Parameters:
    r (float): Base firing rate (spikes per second).
    dt (float): Time step (in seconds).
    total_time (float): Total simulation time (in seconds).
    refractory_period (float): Refractory period (in seconds). Default is 1 ms (0.001 seconds).
    recovery_rate_factor (float): Time constant for linear recovery of the firing rate.
    burst_prob (float): Probability of entering a burst state after a spike. Default is 0 (no bursts).
    burst_duration (float): Average duration of a burst (in seconds). Default is 10 ms (0.01 seconds).
    plot (bool): If True, plot the spike train.

    Returns:
    tuple:
        - np.ndarray: Binary spike train (0 or 1 for each time bin).
        - np.ndarray: Corresponding time vector.
    """
    total_size = int(total_time / dt)  # Number of time bins
    spike_train = np.zeros(total_size, dtype=int)  # Initialize spike train
    time_vector = np.linspace(0, total_time, total_size, endpoint=False)  # Time vector

    # Initialize recovery state
    next_spike_allowed = 0  # Time of next allowed spike due to refractory period
    in_burst = False
    burst_end_time = 0

    for t_idx, t in enumerate(time_vector):
        # Recovery rate: r_recover = r * (t - last_spike) / recovery_rate_factor
        if t < next_spike_allowed:
            recovery_rate = 0  # Refractory period, no spike allowed
        elif in_burst:
            recovery_rate = r * 2  # Higher rate during bursts
        else:
            recovery_rate = r * (t - next_spike_allowed) / recovery_rate_factor

        # Generate a spike based on the current recovery rate
        if np.random.rand() < recovery_rate * dt:
            spike_train[t_idx] = 1  # Spike occurs
            if not in_burst and np.random.rand() < burst_prob:  # Start a burst
                in_burst = True
                burst_end_time = t + burst_duration
            next_spike_allowed = t + refractory_period  # Update refractory window

        # Check if the burst ends
        if in_burst and t >= burst_end_time:
            in_burst = False

    # Plot the spike train
    if plot:
        plt.figure(figsize=(15, 3))
        plt.step(time_vector, spike_train, where='post', color='red')
        plt.xlabel('Time (s)')
        plt.ylabel('Spikes (0 or 1)')
        plt.title(f'Poisson Spike Train with Bursts and Refractory Period')
        plt.show()

    return spike_train, time_vector

# Parameters for fast spiking neuron
r = 80  # Spikes per second
dt = 0.0005  # Time step in seconds (higher resolution)
total_time = 10  # Simulation duration in seconds (shorter for visualization)
refractory_period = 0.0008  # 0.8 ms refractory period
recovery_rate_factor = 2  # Faster recovery
burst_prob = 0.3  # Higher probability of entering a burst
burst_duration = 0.01  # Average burst duration (10 ms)

#bursty neuron activity
spike_train, time_vector = generate_poi_spikes_with_refractory(
    r, dt, total_time, refractory_period=refractory_period, recovery_rate_factor=recovery_rate_factor,
    burst_prob=burst_prob, burst_duration=burst_duration, plot=True
)
```
