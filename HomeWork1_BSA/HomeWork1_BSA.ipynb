### Question 15

\\[
P(\text{Both detect}) = 0.98 \cdot 0.98 = 0.9604.
\\]

Thus:
\\[
P(\text{Both detect}) = 96.04\%.
\\]

#### 2. What is the Probability that at Least One Test Detects a Defect When it is Present?

The probability that at least one test detects a defect is:
\\[
P(\text{At least one detects}) = 1 - P(\text{Neither detects}).
\\]

The probability that neither test detects a defect is:
\\[
P(\text{Neither detects}) = P(\text{Test 1 does not detect}) \cdot P(\text{Test 2 does not detect}).
\\]

Since \( P(\text{Test does not detect}) = 1 - 0.98 = 0.02 \):
\\[
P(\text{Neither detects}) = 0.02 \cdot 0.02 = 0.0004.
\\]

Thus:
\\[
P(\text{At least one detects}) = 1 - 0.0004 = 0.9996.
\\]

The probability that at least one test detects a defect is:
\\[
P(\text{At least one detects}) = 99.96\%.
\\]

#### 3. If the Company Requires Both Tests to Detect a Defect Before Rejecting a Product, Calculate the Probability that a Defective Product Will Be Rejected

If the company requires both tests to detect a defect before rejecting a product, the probability of rejecting a defective product is:

P(\text{Reject defective product}) = P(\text{Both detect}) = 0.9604.
\\]

Thus:
\\[
